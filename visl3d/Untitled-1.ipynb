```python
from jinja2 import Template, Environment, FileSystemLoader

env = Environment(loader=FileSystemLoader('.'), trim_blocks=True, lstrip_blocks=True)
```

```python
temp = env.get_template('javascripts.js.jinja')
```

```python
l_isolevels = [[2,4,6,7,10,12,15,17,20], [1,2,3], [8, 20, 50, 60 ,80]]
iso_split = [1, 5, 1]
```

```python
output = temp.render(l_isolevels=l_isolevels, iso_split=iso_split)
```
