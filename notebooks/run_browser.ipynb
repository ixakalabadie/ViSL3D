```python
import subprocess
import psutil
import time
import pandas as pd
import numpy as np

def open_browser(url):
    subprocess.Popen(['C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe', url])

def monitor_ram(process_name, iter, memory='rss'):
    tot_mean = np.zeros(iter)
    for i in range(iter):
        tab = 0
        tot = 0
        for process in psutil.process_iter(['pid', 'name', 'memory_info']): # _full
            if process.info['name'] == process_name:
                pid = process.info['pid']
                memory_info = process.info['memory_info'] # _full
                if memory == 'rss':
                    mem = memory_info.rss / (1024 ** 2)
                elif memory == 'vms':
                    mem = memory_info.vms / (1024 ** 2)
                elif memory == 'private':
                    mem = memory_info.private / (1024 ** 2)
                print(f"PID: {pid}, RAM Usage: {mem:.2f} MB")
                tab = np.max([tab, mem])
                tot = tot + mem
        print('--------------------------------------')
        print(tot)
        print('--------------------------------------')
        tot_mean[i] = tot
        time.sleep(5)
    return np.mean(tot_mean), tab

def close_browser(process_name):
    for process in psutil.process_iter(['pid', 'name']):
        if process.info['name'] == process_name:
            pid = process.info['pid']
            try:
                # Terminate the process
                psutil.Process(pid).terminate()
                print(f"Browser with PID {pid} terminated.")
            except psutil.NoSuchProcess:
                pass
```

```python
ram_list = []
for (i,names) in enumerate(stats):
     ram_list.append(np.zeros((len(names),2)))
     
     for j in range(len(names)):
        url_to_open = paths[i].replace('rrr',names[j])

        url_to_open = url_to_open + '.html'

        print(url_to_open)
        browser_process_name = "chrome.exe"  # Adjust for the browser you're using (e.g., "firefox.exe" for Firefox)

        open_browser(url_to_open)

        # Allow some time for the browser to open before monitoring
        sl = highno[i][j]/10
        if sl<5:
            sl = 5
        elif highno[i][j] > 700:
            sl = 30
        time.sleep(sl)

        ram = monitor_ram(browser_process_name, 4, memory='rss')
        ram_list[i][j] = ram
        print(ram)

        close_browser(browser_process_name)
```

```python
all_ram = np.concatenate(ram_list)
```

```python
new_pand = pd.DataFrame({'Name': np.concatenate(stats), 'FITS': np.concatenate(fits),
                         'HIcont': np.concatenate(hicont), 'High/No': np.concatenate(highno),
                         'T_RAM': all_ram[:,0], 'P_RAM': all_ram[:,1]})
new_pand.to_csv('stats_RAM_last.csv', index=False)
```
