```python
from visl3d import visual as vis
```

ViSL3D includes the possibility of displaying the visualisations within Jupyter Notebooks in addition to doing it in a browser.<br>
First, we create Cube object the same way as shown in other [example notebooks](https://github.com/ixakalabadie/ViSL3D/blob/master/example/general_example.ipynb).<br>
Then, we run *createVis* but we must set the *notebook* atribute to True. This will create a temporary file that will be opened to visualise it in the cell and removed immediately afterwards.<br>
This way, we can interact with the 3D model directly without needing other software.

```python
cube = vis.prep_one('HCG16.fits', isolevels=[3,4,5,7,10,15,20], unit='rms')
vis.createVis(cube, 'hcg16', notebook=True)
```

ViSL3D also allows displaying previously created (and saved) files in notebooks.<br>
To do that, we just have to input the name of the HTML file we want to visualise to the *displayVis* function as in the example below. In the example we also make the interface minimal.

```python
cube.interface = 'minimal'
vis.displayVis('hcg16_example.html')
```
